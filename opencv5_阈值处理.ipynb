阈值处理

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from torchvision import transforms
```

```python
filename = 'D:\code\pytorch_learn\data\mao.jpg'
```

```python
img = np.array(Image.open(filename))
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.subplot(1, 2, 1); plt.imshow(img, cmap='gray')

t, dst = cv2.threshold(img, 200, 255, type=cv2.THRESH_BINARY)  # 二值化阈值处理
plt.subplot(1, 2, 2); plt.imshow(dst, cmap='gray')
print(t)
print(dst.shape)

plt.show()
```

在进行阈值处理时，自适应阈值处理的方式通过计算每个像素点周围临近区域的加权平均值获得阈值，并使用该阈值对当前像素点进行处理.

```python
fig_size = plt.gcf().get_size_inches()  # 获取当前 figure 尺寸
plt.gcf().set_size_inches(fig_size * 2)   # 放大为两倍

img = np.array(Image.open(filename))
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.subplot(1, 4, 1); plt.imshow(img, cmap='gray')

t1, thd = cv2.threshold(img, 200, 255, cv2.THRESH_BINARY)
plt.subplot(1, 4, 2); plt.imshow(thd, cmap='gray')

athd_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 25, 3)
plt.subplot(1, 4, 3); plt.imshow(athd_mean, cmap='gray')

athd_gaus = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 25, 3)
plt.subplot(1, 4, 4); plt.imshow(athd_gaus, cmap='gray')

plt.show()
```

Otsu方法能够根据当前图像给出最佳的类间分割阈值.

```python
img = np.array(Image.open(filename))
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.subplot(1, 2, 1); plt.imshow(img, cmap='gray')

t, otsu = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
plt.subplot(1, 2, 2); plt.imshow(otsu, cmap='gray')
print(t)

plt.show()
```

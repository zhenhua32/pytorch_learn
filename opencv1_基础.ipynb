````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
````

````python
# 读取图片
filename = 'D:\code\pytorch_learn\data\mao.jpg'
img = cv2.imread(filename, flags=cv2.IMREAD_COLOR)  # 读出来的格式是 BGR
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # 从 BGR 转为 RGB
img = img[..., [2,1,0]]  # 使用 numpy 的方式转换
plt.subplot(1, 2, 1); plt.imshow(img); plt.axis('off')
print(img.shape, type(img))
img = img[:600, ...]
print(img.shape)  # shape 是 (行数, 列数, 通道数)
plt.subplot(1, 2, 2); plt.imshow(img);
# plt.subplot(1, 2, 2); plt.imshow(Image.open(filename)); plt.axis('off')
plt.show()
````

````python
img = np.array(Image.open(filename))
fig, ax = plt.subplots(1, 3, figsize=(10, 10))
ax[0].imshow(img)
a = img[200:400, 200:350]
ax[1].imshow(a)
img[600:800, 400:550] = a
ax[2].imshow(img)
plt.tight_layout()  # 自动调整 padding
plt.show()
````

````python
# 拆分索引
img = np.array(Image.open(filename))
fig_size = plt.gcf().get_size_inches()  # 获取当前 figure 尺寸
print(fig_size)
plt.gcf().set_size_inches(fig_size * 2)   # 放大为两倍
for i in range(3):
    temp = np.full_like(img, 0, dtype=np.uint8)
    temp[:, :, i] = img[:, :, i]
    plt.subplot(1, 3, i+1); plt.imshow(temp); plt.axis('off')
plt.show()
````

````python
img = cv2.imread(filename)
b, g, r = cv2.split(img)  # 通道拆分
print(b.shape)
plt.subplot(1, 3, 1); plt.imshow(b, cmap=plt.cm.Blues_r)
plt.subplot(1, 3, 2); plt.imshow(g, cmap=plt.cm.Greens_r)
plt.subplot(1, 3, 3); plt.imshow(r, cmap=plt.cm.Reds_r)
plt.tight_layout()
plt.show()
````

````python
# 通道合并
img = cv2.imread(filename)
b, g, r = cv2.split(img)
bgr = cv2.merge((b, g, r))
print(bgr.shape)
rgb = cv2.merge((r, g, b))
plt.subplot(1, 2, 1); plt.imshow(rgb)
plt.subplot(1, 2, 2); plt.imshow(bgr)
````
